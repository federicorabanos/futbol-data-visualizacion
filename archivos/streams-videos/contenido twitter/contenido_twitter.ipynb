```python
columnas = [
    'Player',
    'stats_Nation',
    'stats_Pos',
    'stats_Squad',
    'stats_Comp',
    'stats_Age',
    'stats_Born',
    'stats_MP',
    'stats_Starts',
    'stats_Min',
    'stats_90s',
    'stats_xG',
    'stats_npxG',
    'stats_xAG',
    'stats_npxG+xAG',
    'stats_PrgC',
    'stats_PrgP',
    'stats_PrgR',
    'passing_xAG',
    'passing_xA',
    'passing_KP',
    'passing_1/3',
    'passing_PPA',
    'passing_CrsPA',
    'passing_PrgP',
    'gca_SCA',
    'gca_GCA',
    'possession_Mid 3rd',
    'possession_Att 3rd',
    'possession_Att Pen',
    'possession_Att',
    'possession_Succ',
    'possession_Succ%',
    'misc_Int',
    'misc_TklW',
    'misc_Recov',
]

columnas_con_menos_data = [
    'Player',
    'stats_Nation',
    'stats_Pos',
    'stats_Squad',
    'stats_Comp',
    'stats_Age',
    'stats_Born',
    'stats_MP',
    'stats_Starts',
    'stats_Min',
    'stats_90s',
    'passing_KP',
    'passing_1/3',
    'passing_PPA',
    'passing_CrsPA',
    'gca_SCA',
    'gca_GCA',
    'possession_Mid 3rd',
    'possession_Att 3rd',
    'possession_Att Pen',
    'possession_Att',
    'possession_Succ',
    'possession_Succ%',
    'misc_Int',
    'misc_TklW',
]
```

```python
df_total = pd.DataFrame()
for liga in lista_ligas:
    df = pd.read_csv(f'datasets/{liga} - player stats.csv')
    try:
        df = df[columnas]
    except KeyError:
        df = df[columnas_con_menos_data]
    df_total = pd.concat([df_total, df])
```

```python
columnas_90s = [
 'stats_xG',
 'stats_npxG',
 'stats_xAG',
 'stats_npxG+xAG',
 'stats_PrgC',
 'stats_PrgP',
 'stats_PrgR',
 'passing_xAG',
 'passing_xA',
 'passing_KP',
 'passing_1/3',
 'passing_PPA',
 'passing_CrsPA',
 'passing_PrgP',
 'gca_SCA',
 'gca_GCA',
 'possession_Mid 3rd',
 'possession_Att 3rd',
 'possession_Att Pen',
 'possession_Att',
 'possession_Succ',
 'misc_Int',
 'misc_TklW',
 'misc_Recov']
```

```python
for col in columnas_90s:
    df_total[f'{col}_90s'] = df_total[col] / df_total['stats_90s']
```

```python
df_total = df_total.reset_index(drop=True)
```

```python
df_total.columns
```

```python
df_filtrado = df_total[(df_total['passing_KP_90s'] > .8) & (df_total['possession_Succ_90s'] > .6) & (df_total['stats_PrgP_90s'] > 1) & (df_total['stats_Born'] > 1993) & (df_total['stats_90s'] > 3)].reset_index(drop=True)
```

```python
df_filtrado.sort_values(by='possession_Succ_90s', ascending=False)[['Player', 'possession_Succ_90s', 'possession_Succ', 'stats_90s']]
```

```python
fig, ax = plt.subplots(figsize = (16,9))
ejex = 'possession_Succ_90s'
ejey = 'possession_Succ%'
ax.scatter(ejex, ejey, data = df_filtrado)
ax.set_xlabel(ejex)
ax.set_ylabel(ejey)

df_filtrado['zscore'] = stats.zscore(df_filtrado[ejex])*.4 + stats.zscore(abs(df_filtrado[ejey]))*.6
df_filtrado['annotated'] = [True if x > df_filtrado['zscore'].quantile(.9) else False for x in df_filtrado['zscore']]
annotated_df = df_filtrado[df_filtrado['annotated']].reset_index(drop=True)
jugadores = list(annotated_df.Player)
for jugador in jugadores:
    df_jugador = df_filtrado[df_filtrado['Player'] == jugador]
    ax.text(df_jugador[ejex].iloc[0], df_jugador[ejey].iloc[0], s=jugador)
```
